```python
#importing libraries 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
def data_info(dataframe, head=5):

    # boyut bilgisi
    print("\n------------------- Information of Shape -------------------")
    nRow, nCol = dataframe.shape
    print(f'Veri setinde {nRow} satır ve {nCol}  sütun yer almaktadır.')

    #info
    print("\n------------------- Info -------------------")
    print(dataframe.info())

    #Veri setinde herhangi bir eksik değer var mı bakalım
    print("\n------------------- Null or NaN Data ----------------------")
    print(dataframe.isnull().sum())


```

**2.1.2) Kategorik verilerde bar grafikleri ile analizler**

Kategorik değişkenleri incelemek için bar grafiklerini kullandım.

   * a) Tüm katılımcılar

   * b) Madalya alan katılımcılar

```python
def plotBar(df,variable, n=5):
    
    variable_data = df[variable]
    df_count = variable_data.value_counts()
    df_count = df_count[:n]
    
    
    plt.figure()
    plt.bar(df_count.index, df_count, color = ["#"+''.join([random.choice('ABCDEF0123456789') for i in range(6)])])
    plt.xticks(df_count.index, df_count.index.values)
    plt.xticks(rotation=45)
    plt.ylabel("Frekans")
    plt.title("Veri Sıklığı - {}".format(variable))
    plt.show()
    
    print("{}: \n {}".format(variable, df_count))

df_categories_data = df.select_dtypes(include='object').columns
for i in df_categories_data:
    plotBar(df,i, 5)
```

**3 . 1) Madalya kazananların cinsiyete göre yaş boy ve kilo ortalamaları** 

```python
datas_pivot = datas_copy.pivot_table(index=["Medal_Gold","Medal_Silver","Medal_Bronze"], columns="Sex", 
                                values=["Height", "Weight", "Age"], 
                                aggfunc={"Height": np.mean, 
                                       "Weight": np.mean,
                                       "Age":[min,max,np.mean]})

datas_pivot
```

**3.2 ) Spor dallarına göre sporcuların boy, kilo ve yaş ortalamaları**

```python
sport_height_weight = datas_copy.pivot_table(index=["Sport"], columns="Sex", 
                                values=["Height", "Weight", "Age"], 
                                aggfunc={"Height": np.mean, 
                                       "Weight": np.mean,
                                       "Age":[min,max,np.mean]})

sport_height_weight.head(5)
```

**3.3 )İstenilen bir spor dalı için cinsiyetlere göre kaç madalya kazanıldığı bulma**

```python
temp_df = winMedal_df[winMedal_df.Sport=="Aeronautics"]
total = temp_df.Sex.value_counts().reset_index(name = 'Medal')
total.head()
```

```python
def sport_gender(sport):
    plt.figure(figsize=(15,10))
    temp_df = winMedal_df[winMedal_df.Sport==sport]
    total = temp_df.Sex.value_counts().reset_index(name = 'Medal')
    g = sns.catplot(x="index", y="Medal", data = total, height = 10, kind = "bar", palette = "rocket")
    g.despine(left = True)
    g.set_xlabels("Cinsiyet dağılımı")
    g.set_ylabels("Number of medals")
    plt.title(f"{sport} spor dalında alınan madalyalar")
sport_gender("Archery")
```

### 5 ) ZAMAN SERİLERİNDE (TIME SERIES) VERİ ANALİZİ

* 5.1) Yıllara göre ortalama yaş,boy,kilo değişimi
* 5.2) Yıllara göre madalya sayıları

```python
datas_time = winMedal_df.copy()
unique_years = datas_time.Year.unique()
unique_years_sort = np.sort(unique_years)
print(unique_years_sort)
```

Görüldüğü gibi olimpiyatlar çift yıllarda düzenlenmiş

```python
#date time veri tipine dönüştürelim
date_time_obj = pd.to_datetime(datas_time["Year"], format = '%Y')
print(type(date_time_obj))
datas_time["Date_Time"] = date_time_obj # yeni kolon ekledim
datas_time.head(3)
```

```python
#indeksi Date_Time olarak değiştirdim
datas_time = datas_time.set_index("Date_Time") 
datas_time.drop(["Year"], axis=1, inplace = True) 
```

```python
datas_time.head(2)
```

**5.1) Yıllara göre ortalama yaş,boy,kilo değişimi**

```python
# Olimpiyatlar 2 yılda bir olduğu için 2 yıllık periyot aldım
period_data = datas_time.resample("2A").mean() 
period_data.head()
```

Görüldüğü gibi NaN değerler var bunları çıkaralım

```python
period_data.dropna(axis=0, inplace=True)
period_data.head()
```

```python
def difference_years(feature):
    plt.figure()
    period_data[feature].plot()
    plt.title(f"Yıllara göre {feature} Değişimi")
    plt.xlabel("Yıl")
    plt.ylabel(feature)
    plt.grid(True)
    plt.show()

difference_years("Age")
difference_years("Height")
difference_years("Weight")
```

Üç özellik değişimini tek grafikte görmek istersek;

```python
plt.figure()
period_data.plot()
plt.title("Yıllara göre Ortalama Yaş, Boy ve Ağırlık Değişimi")
plt.xlabel("Yıl")
plt.grid(True)
plt.show()
```

**5.2) Yıllara göre madalya sayıları**

```python
datas_year = pd.get_dummies(datas_time, columns=["Medal"])
periods_data = datas_year.resample("2A").sum()
periods_data.head()
```

Bazı yılların değeri sıfır (0), onlardan kurtulalım.

```python
periods_data = periods_data[~(periods_data == 0).any(axis=1)]
periods_data.head()
```

```python
plt.figure()
periods_data.loc[:,["Medal_Bronze", "Medal_Gold", "Medal_Silver"]].plot()
plt.title("Yıllara Göre Madalya Sayıları")
plt.ylabel("Sayı")
plt.xlabel("Yıl")
plt.grid(True)
plt.show()
```
